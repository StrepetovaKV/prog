```python
# Задание 1

def foo(a):
    """Сортировка массива пузырьком.

    Алгоритм: Попарное сравнение и перестановка соседних элементов.

    Параметры:
        - a (list): список чисел.

    Сложность: O(N^2).
    """
    for i in range(len(a), 0, -1):
        for j in range(1, i):
            if a[j-1] > a[j]:
                a[j-1], a[j] = a[j], a[j-1]
    return a

a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python
# Задание 2

def foo(i):
    """Преобразование целого числа в строку.

    Алгоритм: Построение строки путем выделения последней цифры и добавления её в начало результата.

    Параметры:
        - i (int): число.

    Сложность: O(log(N)).
    """
    digits = "0123456789"
    if i == 0:
        return "0"
    result = ""
    while i > 0:
        result = digits[i % 10] + result
        i = i // 10
    return result
```

```python
# Задание 3

def foo(s):
    """Нахождение суммы всех цифр в строке.

    Алгоритм: Перебор всех символов строки с проверкой, является ли символ цифрой, и суммирование цифр.

    Параметры:
        - s (str): строка.

    Сложность: O(N).
    """
    val = 0
    for c in s:
        if c.isdigit():
            val += int(c)
    return val
```

```python
# Задание 4

def foo(n):
    """Поиск всех простых чисел от 1 до n.

    Алгоритм: Перебор всех делителей для каждого числа.

    Параметры:
        - n (int): число.

    Сложность: O(N^2).
    """
    res = []
    for i in range(1, n + 1):
        divisors = 0
        j = 2
        while j < i and divisors == 0:
            if i % j == 0:
                divisors += 1
            j += 1

        if divisors == 0:
            res.append(i)

    return res
```

```python
# Задание 5

def foo(nums):
    """Проверка наличия хотя бы одного чётного числа в списке.

    Алгоритм: Последовательная проверка элементов на чётность.

    Параметры:
        - nums (list): список.

    Сложность: O(N).
    """
    for x in nums:
        if x % 2 == 0:
            return True
    else:
        return False
```

```python
# Задание 8

import random
import time


def timeit(func, *args, **kw):
    """Выполнить функцию 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def maxmin1(data):
    """Вернуть минимум и максимум из 'data'.

    Алгоритм: Последовательный перебор всего массива.

    Сложность: O(N).

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    min_val = max_val = data[0]  # Инициализация минимального и максимального значений
    for num in data[1:]:
        if num < min_val:
            min_val = num
        if num > max_val:
            max_val = num
    return min_val, max_val


def maxmin2(data):
    """Вернуть минимум и максимум из 'data'.

    Алгоритм: Сортировка списка, и возврат первого и последнего элементов.

    Сложность: O(N log N).

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    data_sorted = sorted(data)
    return data_sorted[0], data_sorted[-1]


def maxmin3(data):
    """Вернуть минимум и максимум из 'data'.

    Алгоритм: Используя встроенные функции min() и max().

    Сложность: O(N).

    Параметры:
        - data (list of int): список целых чисел.

    Результат:
        tuple: мин. и макс. значения."""

    return min(data), max(data)


if __name__ == "__main__":
    GEN_LIMIT = 1000000
    dataset = []
    for i in range(4, 7):
        dataset.append(random.sample(range(-GEN_LIMIT, GEN_LIMIT), 10**i))

    res = [["i", "1 способ (мс.)", "2 способ (мс.)", "3 способ (мс.)"]]
    for data in dataset:
        res.append([len(data),
                   timeit(maxmin1, data)[0],
                   timeit(maxmin2, data)[0],
                   timeit(maxmin3, data)[0]])

    # Вывод
    print("{:>15} {:>15} {:>15} {:>15}".format(*res[0]))
    for r in res[1:]:
        print("{:>15} {:>15.2f} {:>15.2f} {:>15.2f}".format(*r))
```

Сравнение:

Способ 1 и способ 3 имеют одинаковую асимптотику O(N), но способ 1 требует больше ручной работы (перебора), в то время как способ 3 просто использует стандартные функции.

Способ 2 имеет наихудшую сложность O(N log N) из-за сортировки, хотя в некоторых случаях может быть быстрее, если сортировка эффективно реализована для маленьких наборов данных.

```python
# Задание 9

import random
import time


def timeit(func, *args, **kw):
    """Выполнить функцию 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def is_ok_to_pass_1(db, passport_id):
    """Вернуть отметку о допуске из 'db' для паспорта с номером 'passport_id'.

    Алгоритм:
        Последовательный перебор списка.

    Сложность: O(N).

    Параметры:
        - db (list of dict): база данных;
        - passport_id (str): номер паспорта.

    Результат:
        bool: отметка о допуске."""

    for record in db:
        if record["id"] == passport_id:
            return record["ok"]
    return False


def is_ok_to_pass_2(db_opt, passport_id):
    """Вернуть отметку о допуске из 'db_opt' для паспорта с номером 'passport_id'.

    Алгоритм:
        Использование словаря для быстрого поиска.

    Сложность: O(1).

    Параметры:
        - db_opt (dict): оптимизированная база данных для поиска;
        - passport_id (str): номер паспорта.

    Результат:
        bool: отметка о допуске."""

    return db_opt.get(passport_id, False)


def ok_str(flag):
    return "Да" if flag else "Нет"


if __name__ == "__main__":
    print("Загрузка базы данных... ")

    GEN_LIMIT = 1000000
    # База данных
    db = [{"id": "{:07d}".format(i), "ok": random.random() < 0.9}
          for i in random.sample(range(GEN_LIMIT, 10 * GEN_LIMIT), GEN_LIMIT)]

    # Оптимизированная база данных (словарь)
    db_opt = {record["id"]: record["ok"] for record in db}

    print("Загружено записей: {:,}.\n\nПримеры:".format(GEN_LIMIT))
    print("Первый элемент:", db[0])
    print("Последний элемент:", db[-1])
    print("\n")

    passport_id = input("Введите номер паспорта (7 цифр): ").strip()

    output = """
                  1 способ   2 способ
    Допущен     {:>10s} {:>10s}
    Время (мс.) {:>10.2f} {:>10.2f}\
    """

    res1 = timeit(is_ok_to_pass_1, db, passport_id)
    res2 = timeit(is_ok_to_pass_2, db_opt, passport_id)

    print(output.format(ok_str(res1[1]),
                        ok_str(res2[1]),
                        res1[0],
                        res2[0]))
```

Сравнение:

Способ 1, несмотря на свою простоту, имеет линейную сложность, что делает его медленным при большом количестве записей.
Способ 2 использует словарь, что значительно ускоряет поиск (константное время) и является гораздо более эффективным для больших баз данных.

```python
# Задание 10

import random
import time


def timeit(func, *args, **kw):
    """Выполнить функцию 'func' с параметрами '*args', '**kw' и
    вернуть время выполнения в мс."""

    time_start = time.time()
    res = func(*args, **kw)
    time_end = time.time()

    return (time_end - time_start) * 1000.0, res


def top3_1(db):
    """Вернуть список из ТОП-3 клиентов по сумме активов.

    Алгоритм:
        Сортировка (встроенный метод sort()) и возврат 3-х элементов.

    Сложность: O(N log N).

    Параметры:
        - db (list of dict): список клиентов.

    Результат:
        tuple of dict: 3 клиента с максимальной суммой активов."""

    db_sorted = sorted(db, key=lambda x: x["assets"], reverse=True)
    return db_sorted[:3]


def top3_2(db):
    """Вернуть список из ТОП-3 клиентов по сумме активов.

    Алгоритм:
        Прямой перебор (1 проход в цикле).

    Сложность: O(N).

    Параметры:
        - db (list of dict): список клиентов.

    Результат:
        tuple of dict: 3 клиента с максимальной суммой активов."""

    top_3 = []
    for client in db:
        if len(top_3) < 3:
            top_3.append(client)
            top_3.sort(key=lambda x: x["assets"], reverse=True)
        else:
            if client["assets"] > top_3[2]["assets"]:
                top_3[2] = client
                top_3.sort(key=lambda x: x["assets"], reverse=True)
    return top_3


if __name__ == "__main__":
    print("Загрузка базы данных... ")

    GEN_LIMIT = 10000000
    # Список клиентов
    db = []
    for client_id, assets in enumerate(
      random.sample(range(0, 10 * GEN_LIMIT), GEN_LIMIT),
      start=1):
        db.append(dict(client_id=client_id, assets=assets))

    print("Загружено записей: {:,}.\n\nПримеры:".format(GEN_LIMIT))
    print("Первый элемент:", db[0])
    print("Последний элемент:", db[-1])
    print("\n")

    res1 = timeit(top3_1, db)
    res2 = timeit(top3_2, db)

    for i, (func_time, res) in enumerate((res1, res2), start=1):
        print("Способ №{}".format(i))
        print("Время: {:.2f} мс.".format(func_time))
        for j, client in enumerate(res, start=1):
            print("{}-е место: ID={client_id} Активы={assets:,} руб.".
                  format(j, **client))
        print()
```

Сравнение:

Способ 1 имеет более высокую асимптотическую сложность (O(N log N)), так как сортировка требует больше времени, особенно при большом числе клиентов. Однако этот способ может быть проще в реализации и предоставляет полностью отсортированный список.
Способ 2 — это оптимизация, которая работает за линейное время O(N), но требует немного более сложной логики. Это идеальный выбор, когда необходимо получить только топ-3 элементов, не сортируя весь список.
